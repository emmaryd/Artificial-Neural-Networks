```python
import numpy as np

class Hopfield():
    """
    Creates a Hopfield Network.
    """
    def __init__(self):
        self.weights = None

    def hebbs_rule(self, stored_patterns, diagonal_set_to_zero=True):
        """
        Generates the weight matrix for the stored patterns.
        """
            
        num_bits = len(stored_patterns[0]) 
        self.weights = (1/num_bits)* np.dot(stored_patterns.T,
                                            stored_patterns) 
        if diagonal_set_to_zero == True:
            for k in range(len(stored_patterns[0])):
                self.weights[k][k] = 0
            
        return self.weights
    
    def deterministic_asynchronous(self, index_neuron, feed_pattern): 
        """
        Makes one asynchronous update to a choosen neuron in one fed pattern.
        """
        
        original_neuron = feed_pattern[index_neuron]
        updated_neuron = int(np.sign(np.dot(self.weights[index_neuron].T,
                                            feed_pattern.T)))

        feed_pattern[index_neuron] = updated_neuron
        
        return original_neuron, updated_neuron, feed_pattern
```
